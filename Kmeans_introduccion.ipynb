```python
#calculo de distancia euclidiana:
from scipy.spatial.distance import euclidean

point1=[1,2]
point2=[4,6]
distance=euclidean(point1,point2)

print('Distancia eiclidiana entre los puntos:' ,distance)
```

```python
import math
distance = math.sqrt((point2[0] - point1[0])**2 + (point2[1] - point1[1])**2)

print('Distancia euclidiana entre los puntos:', distance)
```

La distancia de Manhattan (también conocida como distancia L1 o distancia de taxista) entre dos puntos \( A = (x_1, y_1) \) y \( B = (x_2, y_2) \) se calcula como:

\[ \text{distancia} = |x_2 - x_1| + |y_2 - y_1| \]

```python
#distancia manhatan

distance = abs(point2[0] - point1[0]) + abs(point2[1] - point1[1])

print('Distancia de Manhattan entre los puntos:', distance)
```

```python
#Implementacion de K-Mean

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#generacion de datos aleatorios para clustering
X=np.random.rand(100,2)


#Inicializar y ajusta r el modelo K-Means
kmeans=KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
#vidualizazcion de los cluster
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1],marker='x',color='red',s=200);

plt.title('Clustering con K-Means')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
```

```python
# cargar conjunto de datos Iris

iris=load_iris()
X=iris.data
y=iris.target

#Normalizar los datos
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
#reducir la dimensionaidad con PCA para visualizacion
pca=PCA(n_components=2)
x_pca=pca.fit_transform(X_scaled)
```

```python
#   Iniciarlizar y ajustar el modelo K-means

kmeans=KMeans(n_clusters=3)
kmeans.fit(X_scaled)

```

```python
#visualizar los cluster

plt.scatter(x_pca[:, 0], x_pca[:, 1], c=kmeans.labels_, cmap='viridis')


plt.title('Clustering con K-Means en conjunto de datos iris')
plt.xlabel('Componente principal   1')
plt.ylabel('Componente principal 2')
plt.show()
```
