# A more complete example with synthetic data given by ChatGPT

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Datos sintéticos para el ejemplo
data = {
    'Número de Hombres': [15, 12, 10, 17, 14, 19, 11, 13, 16, 20],
    'Número de Mujeres': [20, 18, 22, 15, 16, 14, 21, 17, 19, 18],
    'Capacidad del Salón': [35, 30, 32, 40, 33, 40, 32, 30, 35, 38]
}


# Crear el DataFrame
df = pd.DataFrame(data)


correlation_matrix = df.corr()


print("Matriz de Correlación:")
print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap de la Matriz de Correlación')
plt.show()
```

```python
#Matriz de confusión:

'''Es una herramienta que se utiliza en el campo de la estadística y el aprendizaje automático para evaluar el rendimiento de un modelo de clasificación.
Esta matriz muestra la relación entre las predicciones del modelo y los valores reales de las clases.
Es especialmente útil para entender cómo de preciso es un modelo y para identificar tipos específicos de errores.'''

#En un problema de clasificacion Binaria se tendría:
'''
	               Predicción Positiva	      Predicción Negativa
Clase Positiva	 Verdadero Positivo(TP)   	Falso Negativo (FN)
Clase Negativa	 Falso Positivo (FP)	      Verdadero Negativo (TN)'''


#Verdadero Positivo (TP): El número de casos positivos que fueron correctamente predichos como positivos.
#Falso Negativo (FN): El número de casos positivos que fueron incorrectamente predichos como negativos.
#Falso Positivo (FP): El número de casos negativos que fueron incorrectamente predichos como positivos.
#Verdadero Negativo (TN): El número de casos negativos que fueron correctamente predichos como negativos.
```

```python
#Consultar las metricas de desempeño del modelo para clasificación y regresión:

#Regresión:

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_squared_log_error

# Supongamos que tienes dos listas: y_true (valores reales) y y_pred (valores predichos)
y_true = [3, -0, 2, 7]
y_pred = [2.5, 0.0, 2, 8]


#  Error Cuadrático Medio (MSE - Mean Squared Error) Calcular MSE-->El MSE mide el promedio de los errores al cuadrado entre los valores reales y los valores predichos.
mse = mean_squared_error(y_true, y_pred)
print(f"MSE: {mse}")


#  Raíz del Error Cuadrático Medio   Calcular RMSE  --> El RMSE es la raíz cuadrada del MSE y proporciona una medida del error en las mismas unidades que la variable de respuesta.
rmse = mean_squared_error(y_true, y_pred, squared=False)
print(f"RMSE: {rmse}")


# Error Absoluto Medio  Calcular MAE  -->El MAE mide el promedio de los errores absolutos entre los valores reales y los valores predichos.
mae = mean_absolute_error(y_true, y_pred)
print(f"MAE: {mae}")


# Calcular R^2  -- >  mide la proporción de la varianza en la variable dependiente que es predecible a partir de las variables independientes.
r2 = r2_score(y_true, y_pred)
print(f"R^2: {r2}")

# Error Cuadrático Medio de Logaritmos  Calcular MSLE  -- > El MSLE mide el promedio de los errores al cuadrado entre los logaritmos de los valores reales y los valores predichos.
msle = mean_squared_log_error(y_true, y_pred)
print(f"MSLE: {msle}")

#Error Absoluto Medio de Porcentaje  Calcular MAPE  -- > El MAPE mide el promedio de los errores porcentuales absolutos entre los valores reales y los valores predichos.
mape = np.mean(np.abs((np.array(y_true) - np.array(y_pred)) / np.array(y_true))) * 100
print(f"MAPE: {mape}%")




```

```python
#Clásificación:

# (Accuracy): Proporción de predicciones correctas (TP + TN) sobre el total de predicciones.

#Sensibilidad (Recall) o Tasa de Verdaderos Positivos: Proporción de verdaderos positivos sobre el total de casos positivos.     TP/(TP+FN)

#Precison: Proporción de verdaderos positivos sobre el total de predicciones positivas. TP/(TP+FP)

#untaje F1 (F1 Score): La media armónica de la precisión y la sensibilidad:  2* precision*sensibilidad


from sklearn.metrics import classification_report, confusion_matrix

# Supongamos que tienes tus valores reales (y_true) y tus predicciones del modelo (y_pred)
y_true = [0, 1, 1, 0, 1, 1, 0, 0, 1, 0]
y_pred = [0, 1, 0, 0, 1, 1, 1, 0, 1, 0]


conf_matrix = confusion_matrix(y_true, y_pred)
print("Matriz de confusión:")
print(conf_matrix)


report = classification_report(y_true, y_pred, target_names=['Clase 0', 'Clase 1'])
print("\nInforme de clasificación:")
print(report)

```
